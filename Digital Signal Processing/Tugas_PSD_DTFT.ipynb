```
import numpy as np
import matplotlib.pyplot as plt
```

```
def plot_freq_domain(X,freq='half',title=''):
  '''
  Fungsi untuk menampilkan tanggap frekuensi
  input:
    X : hasil transformasi fourier
    freq : jangkauan frekuensi radian yg ditampilkan
      'half' : 0 - pi
      'full' : 0 - 2pi
    title : judul fungsi 
  '''
  N = len(X)
  if freq=='half':
    N_plot = N//2
    X_plot = X[:N_plot]
    w_pi = np.arange(N_plot)/(N_plot)
  elif freq=='full':
    N_plot = N
    X_plot = X
    w_pi = np.arange(N_plot)/(N_plot/2)

  plt.figure(1)
  plt.plot(w_pi,abs(X_plot))
  plt.xlabel(r'$\omega \ (\pi \ rad)$')
  plt.ylabel('magnitudo')
  plt.grid()
  plt.title('Tanggap Frekuensi '+title)

  # plot tanggap fase
  plt.figure(2)
  plt.plot(w_pi,np.angle(X_plot)/np.pi)
  plt.xlabel(r'$\omega \ (\pi \ rad)$')
  plt.ylabel('fase ' + r'$(\pi \ rad)$')
  plt.title('Phase '+title)
  plt.grid()

def DTFT(x,N_w=1000,plot=True,freq='full',title=''):
  '''
  Fungsi untuk menghitung hasil DTFT dengan menggunakan persamaan dasar DTFT
  input:
    x : sinyal x(n) yang akan ditransformasi
    N_w : banyaknya titik untuk indeks frekuensi
    plot : True untuk menampilkan grafik DTFT
           False untuk tidak menampilkan grafik DTFT
    freq : jangkauan frekuensi radian yg ditampilkan, jika plot == True
      'half' : 0 - pi
      'full' : 0 - 2pi
    title : judul fungsi, jika plot == True
  '''
  X = np.zeros(N_w,dtype='complex')
  w = np.arange(N_w)*np.pi/(N_w/2)
  N_x = len(x)
  
  for n in range(N_x):
    X += x[n]*np.exp(-1j*w*n)
  
  if plot:
    plot_freq_domain(X,freq=freq,title=title)

  return X
```
